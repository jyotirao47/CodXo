```python
df_marge = pd.concat([df_false, df_true], axis=0)
df_marge.head(10)
```

```python
df = df_marge.drop(["title","subject", "date"], axis=1)
df.head(10)
```

## **Randomly shuffling the dataframe**

```python
df = df.sample(frac=1)
df.head(10)
```

```python
df.isnull().sum()
```

## **Preprocessing Text**

**Creating a function to convert the text in lowercase, remove the extra space, special chr., ulr and links.bold text**

```python
def word_drop(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)  # Remove text in square brackets
    text = re.sub("\\W", " ", text)  # Remove non-word characters
    text = re.sub('https?://\S+|www\.\S+', ' ', text)  # Remove URLs
    text = re.sub('<.*?>+', '', text)  # Remove HTML tags
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)  # Remove punctuation
    text = re.sub('\n', '', text)  # Remove newlines
    text = re.sub('\w*\d\w*', '', text)  # Remove words containing numbers
    return text
```

```python
df["text"] = df["text"].apply(word_drop)
```

```python
df.head(10)
```

**Defining dependent and independent variable as x and y**

```python
x = df["text"]
y = df["class"]
```

## **Training the model**

**Splitting the dataset into training set and testing set.**

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = .25)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### **Feature Extraction (TF-IDF)**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TfidfVectorizer
vectorization = TfidfVectorizer()

# Fit and transform the training data
xv_train = vectorization.fit_transform(x_train)

# Transform the test data
xv_test = vectorization.transform(x_test)
```

### **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
LR.score(xv_test, y_test)
```

```python
pred_LR = LR.predict(xv_test)
```

```python
print(classification_report(y_test, pred_LR))
```

### **Decision Tree Classification**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
DT.score(xv_test, y_test)
```

```python
pred_DT = DT.predict(xv_test)
print(classification_report(y_test, pred_DT))
```

### **Gradient Boosting Classifier**

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
GBC = GradientBoostingClassifier(random_state=0)
GBC.fit(xv_train, y_train)
```

```python
GBC.score(xv_test, y_test)
```

```python
pred_GBC =GBC.predict(xv_test)
print(classification_report(y_test, pred_GBC))
```

### **Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(random_state=0)
RFC.fit(xv_train, y_train)
```

```python
RFC.score(xv_test, y_test)
pred_RFC = RFC.predict(xv_test)
print(classification_report(y_test, pred_RFC))
```

### **Passive Aggressive Classifier**

```python
from sklearn.linear_model import PassiveAggressiveClassifier
Pac = PassiveAggressiveClassifier(max_iter=50)
Pac.fit(xv_train, y_train)
```

### **Manual Testing**

```python
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["test"] = new_def_test["text"].apply(word_drop)
    new_x_test = new_def_test["test"]
    new_xv_test = vectorization.transform(new_x_test)

    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GBC = GBC.predict(new_xv_test)
    pred_RFC = RFC.predict(new_xv_test)

    print("\n\nLR Prediction: {}\nDT Prediction: {}\nGBC Prediction: {}\nRFC Prediction: {}".format(
        output_label(pred_LR[0]), output_label(pred_DT[0]), output_label(pred_GBC[0]), output_label(pred_RFC[0])
    ))

```
